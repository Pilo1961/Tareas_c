```c
//%cflags:-lm
#include<stdio.h>
#include<stdlib.h>
#include<math.h>

double fl_c_k(double a, int k){
    int signo=1, exp=0;
    double mantisa=0;
    
    if(a<0){
        signo=-1;
        a=a*signo;
    }
    exp=floor(log10(a))+1;
    mantisa= a * pow(10,-exp);
    
    
    
    return signo*pow(10,exp)*(double)((int)(pow(10,k)*mantisa)*pow(10,-k));
}

double err_abs(double aprox, double obj){   
    return fabs(obj-aprox);
}

double err_rel(double errabs, double obj){
    return errabs/fabs(obj);
}

void suma(double x, double y){
    int k;
    double res1=0, res2=0;

    k=8;
    res1 = fl_c_k(fl_c_k(x,k)+fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x+y a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res2 = fl_c_k(fl_c_k(x,k)+fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x+y a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
}


void resta(double x, double y){
    int k;
    double res1=0, res2=0;

    k=8;
    res1 = fl_c_k(fl_c_k(x,k)-fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x-y a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res2 = fl_c_k(fl_c_k(x,k)-fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x-y a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
}

void multiplicacion(double x, double y){
    int k;
    double res1=0, res2=0;

    k=8;
    res1 = fl_c_k(fl_c_k(x,k)*fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x*y a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res2 = fl_c_k(fl_c_k(x,k)*fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x*y a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
}

void division(double x, double y){
    int k;
    double res1=0, res2=0;

    k=8;
    res1 = fl_c_k(fl_c_k(x,k)/fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x/y a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res2 = fl_c_k(fl_c_k(x,k)/fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("x/y a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
}

double resta_dbl(double x, double y, int k){
    double res1=0;
    res1 = fl_c_k(fl_c_k(x,k)-fl_c_k(y,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    return res1;
}

void rest_div(double x, double u, double w){
    int k;
    double res1=0, res2=0, res=0;

    k=8;
    res=resta_dbl(x,u,k);
    res1 = fl_c_k(fl_c_k(res,k)/fl_c_k(w,k) , k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("(x-u)/w a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res=resta_dbl(x,u,k);
    res2 = fl_c_k(fl_c_k(res,k)/fl_c_k(w,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("(x-u)/w a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
    
}

void rest_mult(double x, double u, double v){
    int k;
    double res1=0, res2=0, res=0;

    k=8;
    res=resta_dbl(x,u,k);
    res1 = fl_c_k(fl_c_k(res,k)*fl_c_k(v,k) , k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("(x-u)*v a %d dígitos %0.7e\n", k, res1);
    
    k=5; //precisión 5
    res=resta_dbl(x,u,k);
    res2 = fl_c_k(fl_c_k(res,k)*fl_c_k(v,k),k-1); //k-1 para seguir manteniendo 5 dígitos pues la suma aumenta un dígito
    printf("(x-u)*v a 5 dígitos %0.4e\n", res2); 
    
    printf("Error absoluto de aprox: %e\n",err_abs(res2,res1));
    printf("Error relativo de aprox: %e\n",err_rel(err_abs(res2,res1),res1));
    
}

int main(){
    double x = 5/7.0;
    double y = 1/3.0;
    double u= 0.714251;
    double v= .987659E5;
    double w=.111111E-4;
    double res1=0;
    double res2=0;
    int k;

    k=8; //precisión 8
    printf("precision 8\n");
    printf("valor x a 8 dígitos: %0.7e\n", fl_c_k(x,k));
    printf("valor y a 8 dígitos: %0.7e\n", fl_c_k(y,k));
    printf("valor u a 8 dígitos: %0.7e\n", fl_c_k(u,k));
    printf("valor v a 8 dígitos: %0.7e\n", fl_c_k(v,k));
    printf("valor w a 8 dígitos: %0.7e\n", fl_c_k(w,k));
    
    
    k=5; //precisión 5
    printf("precision 5\n");
    printf("valor x a 5 dígitos: %0.4e\n", fl_c_k(x,k));
    printf("valor y a 5 dígitos: %0.4e\n", fl_c_k(y,k));
    printf("valor u a 5 dígitos: %0.4e\n", fl_c_k(u,k));
    printf("valor v a 5 dígitos: %0.4e\n", fl_c_k(v,k));
    printf("valor w a 5 dígitos: %0.4e\n", fl_c_k(w,k));
    
    printf("--------------------------\n");
    printf("SUMA\n");
    suma(x,y);
    
    printf("--------------------------\n");
    printf("RESTA\n");
    resta(x,y);
    
    printf("--------------------------\n");
    printf("MULTIPLICACION\n");
    multiplicacion(x,y);
    
    printf("--------------------------\n");
    printf("DIVISION\n");
    division(x,y);
    
    printf("--------------------------\n");
    printf("RESTA x-u\n");
    resta(x,u);
    
    printf("--------------------------\n");
    printf("(x-u)/w  \n");
    rest_div(x,u,w);
    
    printf("--------------------------\n");
    printf("(x-u)*v \n");
    rest_mult(x,u,v);
    
    printf("--------------------------\n");
    printf("suma u,v \n");
    resta(u,v);
    
    
    return 0;
}

```
